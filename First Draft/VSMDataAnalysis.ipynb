```python
import pandas as pd
import numpy as np
from pandas.compat import StringIO
import matplotlib.pyplot as plt
```

```python
Pdata = pd.read_csv('VSM/powderVSM.csv',delimiter = ',')
xStr = 'Temperature (K)'
yStr = 'M. Raw\' (emu)'
eStr = 'M. Std. Err. (emu)'
#xStr =  'Magnetic Field (Oe)'
# other interesting parameter includes
# M. Raw" (emu)
# Moment (emu)
# M. Quad. Signal (emu)


VSM = Pdata[[xStr,yStr,eStr]]
VSM1 = VSM.iloc[0:70,:]
ax1 = VSM1.plot(xStr,yStr, figsize=(20,10))
ax1.set_ylim(-0.000003,0.000003)
ax1.set_title('Vibrating Magnetic Sample Measurement for YBCO powder', fontsize=20)
ax1.errorbar(VSM1[[xStr]],VSM1[[yStr]],fmt='o')
ax1.grid(True)
```
